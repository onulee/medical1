```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
# (국어+영어+수학+과학+사회) -> 총합,평균의 컬럼을 추가하고
# 평점이라는 컬럼을 만들어서, 평균 90점이상이면 A, 80점 B, C, D, 60점미만 F 함수를 사용해서 처리하시오.

df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df['평균'] = df['총합'] / 5
```

```python
def grade(score):
    if score >= 90 : return "A"
    elif score >= 80 : return "B"
    elif score >= 70 : return "C"
    elif score >= 60 : return "D"
    else: return "F"
```

```python
df['평점'] = df['평균'].apply(grade)
```

```python
def result(score):
    txt = ''
    if score >= 70 : txt = "합격"
    else: txt = "불합격"
    return txt
```

lambda함수 : 익명함수 -> lambda 매개변수

```python
# lambda함수 : 익명함수 -> lambda 매개변수
# 결과 컬럼 평균 70점이상 인 학생은 합격, 불합격
# df['결과'] = df['평균'].apply(result)
df['결과'] = df['평균'].apply(lambda score:'합격' if score>=70 else "불합격")
```

```python
### 결과2 컬럼을 생성 65점 초과 합격, 65점 재시험, 65점미만 불합격
df['결과2'] = df['평균'].apply(lambda score:'합격' if score>65 else ( "재시험" if score==65 else "불합격" ))
```

```python
def lower_change(lang):
    # lang null 아니냐? , notnull : null이 아닌지 확인, isnull : null인지 확인
    if pd.notnull(lang):  return lang.lower()
    else: return "python"
```
