```python
import random
df['요일'] = 0
```

```python
# 0-6까지의 숫자를 랜덤으로 입력
for i in range(8):
    df.iloc[i,9] = random.randint(0,6)
```

```python
def change(x):
    if x==0: return "일요일"
    elif x==1 : return "월요일"
    elif x==2 : return "화요일"
    elif x==3 : return "수요일"
    elif x==4 : return "목요일"
    elif x==5 : return "금요일"
    elif x==6 : return "토요일"
```

```python
# 0 - 일요일 1,2,3,4,5,6-토요일
# 요일 컬럼에 숫자를 요일로 변경해서 출력하시오.
df['요일'] = df['요일'].apply(change)
df
```
