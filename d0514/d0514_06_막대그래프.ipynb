```python
# score.xlsx
# 합계 컬럼을 생성해서 막대그래프, 옆으로 누운 그래프 를 출력하시오.
# 합계점수를 출력할것 
# 국어+영어+수학+과학+사회
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['axes.unicode_minus'] = False  # 마이너스 표시 해결
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # windows 사용자
# matplotlib.rcParams['font.family'] = 'AppleGothic Gothic' # Mac사용자
matplotlib.rcParams['font.size'] = '10' # 글자크기
```

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df.head(3)
```

```python
x = df['이름']
df['합계'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
y = df['합계']
```

```python
# 키를 막대 그래프로 출력하시오.
plt.bar(x,y,label='합계')
plt.legend(loc='upper center')  # 위치지점 변경

for i,txt in enumerate(y):
    plt.text(x[i],y[i]+2.0,txt,ha='center',color='blue')

# plt.xticks([165,170,180,190,200,210])    # 그래프 y축 눈금표시
plt.ylim(0,500) # 그래프 눈금 범위 지정
plt.title('합계 그래프')
plt.show()
```

```python
# 키를 막대 그래프로 출력하시오.
plt.barh(x,y,label='합계')
plt.legend(loc='center right')  # 위치지점 변경

for i,txt in enumerate(y):
    plt.text(y[i]+12.5,x[i],txt,ha='center',color='blue')

# plt.xticks([165,170,180,190,200,210])    # 그래프 y축 눈금표시
plt.xlim(0,500) # 그래프 눈금 범위 지정
plt.title('합계 그래프')
plt.show()
```
