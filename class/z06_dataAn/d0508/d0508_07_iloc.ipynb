```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
## loc index명, column명을 가지고 검색
df.loc['1번':'5번','국어':'사회']

## iloc 번호를 가지고 검색
```

### iloc출력

```python
df.iloc[[2,5,6]]
df
```

```python
# 5번,8번 학생출력
df.iloc[[4,7]]
# 4번학생 키 출력
df.iloc[3,2] 

# 4,6번학생의 국어,사회를 출력하시오.
df.iloc[[3,5],[3,7]]
# df.loc[['4번','6번'],['국어','사회']]

# 마지막 row출력
df.iloc[-1]
df.iloc[-4:]

df.iloc[2:4]
# iloc 슬라이싱 2,3
df.iloc[2:4,0:5]

# loc 슬라이싱으로 출력
df.loc['3번':'4번','이름':'영어']
```
