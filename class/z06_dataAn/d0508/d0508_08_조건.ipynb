```python
# a_list = [1,3,5,7]
a_list = [i*2 for i in range(0,4)] # 리스트의 번호는 iloc
a_list
df.iloc[a_list]
filt  = df['키']>188  #조건식 loc
filt
df.loc[filt]
```

```python
# pandas 조건문 - 컬럼의 조건을 부여해서 출력시킴.
df[df['키']>190]
```

```python
df.loc[['1번','6번']]
df.iloc[[0,5]]
```

```python
## 키가 평균보다 큰 학생, 이름,키만 출력하시오.
filt = df['키'] >= df['키'].mean()
df.loc[filt,['이름','키']]
df.loc[~filt,['이름','키']] # ~ 조건에 반대
```

```python
# 조건 키가 170 보다 작거나, 200보다 큰 학생을 출력하시오.
filt = (df['키']<170) | (df['키']>200)
filt
df.loc[filt]
# df[filt]
# df.loc[filt,'키']

# df[0:4]
# df.head()
# df.tail()
```

```python
## 국어 85점이상, 영어 85점이상 이름,학교,국어,영어 출력하시오.
## and & , or |
## and, &&, &  or | ||
## & , | pandas
## 파이썬 and, or

filt = (df['국어'] >= 85) & (df['영어'] >=85)
df.loc[filt,['이름','국어','영어']]
```
