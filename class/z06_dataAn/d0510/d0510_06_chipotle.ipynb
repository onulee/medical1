```python
import pandas as pd
df = pd.read_csv('../z20_data/chipotle.tsv',sep='\t')
df['item_price'] = df['item_price'].str.replace("$","").astype(float)
df
```

```python
# 1. 메뉴별 주문 수량을 출력하시오.
# value_counts : 컬럼 데이터별 개수를 출력
df['item_name'].value_counts()[:10]

# sort_values : 컬럼 데이터 정렬
# df.groupby('item_name').count().sort_values('order_id')['order_id']
# df.groupby('item_name').count().sort_values('order_id')[:10]
# df.groupby('item_name').count().sort_values('order_id',ascending=False)['order_id'][:10]
```

```python
df['item_name'].value_counts()[0] # 개수출력
df['item_name'].value_counts().index[0] # 메뉴명 출력
```

```python
# items() 

list(df['item_name'].value_counts()[:10].items())

item_count = df['item_name'].value_counts()
# items() : 항목과 개수가 튜플형태로 구성
for i,(val,cnt) in enumerate(item_count.items()):
    print("Top",(i+1),":", val,",",cnt)
```

```python
# 2. 주문당 평균 계산금액 출력하시오.
# item당 주문 개수 출력
order_count = df.groupby('item_name')['order_id'].count()
order_count
```

```python
# item당 주문 총량 출력
len(df.groupby('item_name'))
df.groupby('item_name')['quantity'].sum().sort_values(ascending=False)
```

```python
# unique() - 중복제거 : 50개
len(df['item_name'].unique())
# 총 주문수량 : 1832
df['order_id'].unique()
len(df['order_id'].unique())
```

```python
# 각 아이템별 가격을 구하시오.
# 아이템별 주문개수가 1개 인 것의 가격을 출력하시오.
df_one_quantity = df[df['quantity']==1]

# 50개별 최저 가격 - 아이템별 1개 주문 금액
df_item_price = df_one_quantity.groupby('item_name').min().sort_values('item_price')

# 메뉴별 1개 주문 금액
df_item_price['item_price']

# 메뉴개수 : 50
# len(df_one_quantity.groupby('item_name').min().sort_values('item_price')['item_price'])
```

```python
## 주문별, (item_price)가 높은 가격순으로 주문개수를 출력하시오.
# 1. 주문별로 그룹화 한 후 가격 역순으로 정렬하시오.
# 주문개수

df_price_group = df.groupby('order_id').sum().sort_values('item_price',ascending=False)
df_price_group[['quantity','item_price']]
```

```python
# 2. 주문당 평균 계산금액 출력하시오.
# $표시 제거 및 float타입 변경
df['item_price'] = df['item_price'].str.replace("$","").astype(float)
df.groupby('order_id')['item_price'].sum().mean()
```

```python
df.groupby('order_id').size()
len(df.groupby('order_id').size())
```

```python
# 한 주문에 20달러 이상 사용한 주문의 id들 출력하시오.
order_id_group = df.groupby('order_id').sum()
df.groupby('order_id').sum()['item_price']>20
# order_id_group[order_id_group['item_price']>20]

# row에 해당하는 컬럼을 출력할때 loc 입력
# kor_group   .loc[df.groupby('학교')['국어'].sum()>240          ,'국어']
order_id_group.loc[df.groupby('order_id').sum()['item_price']>20,['item_price']]
```

```python
### 주문당 개수가 가장 많은 상위 5개를 출력하시오.
df.groupby('order_id').sum().sort_values('quantity',ascending=False)[['quantity','item_price']][:5]
```
