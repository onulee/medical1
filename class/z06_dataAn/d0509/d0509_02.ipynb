```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
# 키가 189이상인 학생을 출력하시오.
df['키'] > 189
df.loc[~(df['키']>189),['이름','키']]
```

```python
# 키 180 이상이면서 구로고 학생을 출력하시오.
# df['키'] >= 180
df['학교'] == '구로고'

df[ (df['키'] >= 180) & (df['학교'] == '구로고') ]
```
