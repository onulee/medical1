```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
import numpy as np
df['학교'] = np.nan
```

#### dropna - Nan처리 : 지우기
axis : index,colums
how : any,all

```python
# Nan 가 들어가 있는 컬럼을 전체 삭제
df.dropna(axis='columns', inplace=True)
```

```python
# Nan 들어가 있는 컬럼에서 모든 컬럼이 Nan 인것만 삭제
df.dropna(axis='columns', how='all', inplace=True)
```

```python
# Nan 들어가 있는 컬럼에서 1개라 Nan이 포함되어 있으면 삭제
df.dropna(axis='columns',how='any',inplace=True)
```

```python
# Nan 들어가 있는 index에서 1개라도 Nan이 포함되어 있으면 삭제
df.dropna(axis='index',how='any',inplace=True)
```
