```python
import pandas as pd
data = {
   '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
   '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
   '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
   '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
df
```

```python
# 관객수 전체평균 관객수 보다 높은 영화를 출력하시오.
# 전체 평균
df['관객 수'].mean()
```

```python
df[df['관객 수']>=df['관객 수'].mean()]
```

```python
# 2015년 이후 개봉한 영화 데이터에서 영화,개봉연도 만 출력하시오.
# 1. 2015년 개봉한 영화
# df['개봉 연도']>=2015
df.loc[df['개봉 연도']>=2015,['영화','개봉 연도']]
# df[df['개봉 연도']>=2015]
```

```python
df
```

```python
## (관객수 * 평점) // 100 을 적용해서, 추천 점수 컬럼을 생성하시오. 추천점수 컬럼 역순정렬
df['추천 점수'] = df['관객 수'] * df['평점'] // 100
```

```python
df.sort_values('추천 점수',ascending=False)
```

```python
# 개봉연도 역순정렬 출력
df.sort_values('개봉 연도',ascending=False)
```

```python
import pandas as pd
data = {
   '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
   '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
   '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
   '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
df
```

```python
def change(grade):
    return str(grade)+'점'
```

```python
df['평점'] = df['평점'].apply(change)
```
