```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['axes.unicode_minus'] = False  # 마이너스 표시 해결
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # windows 사용자
# matplotlib.rcParams['font.family'] = 'AppleGothic Gothic' # Mac사용자
# matplotlib.rcParams['font.size'] = '15' # 글자크기
```

```python
x = [1,2,3]
y = [12,9,8]

# 그래프 그리기, 선두께 linewidth, label:범례
# plt.plot(x,y,linewidth=3,marker='o',markersize=10,markeredgecolor='blue', markerfacecolor='yellow',label='국어')
# markersize -> ms 마커 크기
# markedgecolor -> mec 마커 테두리색상
# markerfacecolor -> mfc 마커 색상
# linestyle -> ls 선 스타일
# alpha 투명도
plt.plot(x,y,linewidth=3,marker='o',ms=20,mec='red', mfc='yellow',label='국어',alpha=0.7,ls=':')
# plt.legend(loc='upper right')
plt.legend(loc=(0.2,0.5))  # 위치지점 변경
plt.title('막대그래프')
# x,y축 표시 간격을 조정
plt.yticks([5,7,9,11,13,15])
plt.xticks([1,2,3])
```
