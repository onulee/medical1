```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['axes.unicode_minus'] = False  # 마이너스 표시 해결
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # windows 사용자
# matplotlib.rcParams['font.family'] = 'AppleGothic Gothic' # Mac사용자
matplotlib.rcParams['font.size'] = '10' # 글자크기

x = [1,2,3]
y = [2,4,8]
```

```python
# plt.plot(x,y,label='국어')
# 라인 스타일
plt.plot(x,y,label='국어',linestyle='--',marker='o')
plt.legend(loc='upper center')
plt.show()
```

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df.head(3)
```

```python
x = df['이름']
y = df['키']
```

```python
# 키를 꺽은 선 그래프로 출력하시오.
# 이름, 키
# y축표시 170,180,190,200,210
plt.plot(x,y,linewidth=3,marker='o',ms=10,mec='blue', mfc='yellow',label='키')
plt.legend(loc='upper center')  # 위치지점 변경

for i,txt in enumerate(y):
    plt.text(x[i],y[i]+1.8,txt,ha='center',color='blue')

# plt.yticks([150,160,170,180,190,200,210])    # 그래프 y축 눈금표시
plt.ylim(165,210) # 그래프 눈금 범위 지정
plt.title('키 그래프')
```

```python
# 키를 막대 그래프로 출력하시오.
# 이름, 키
# y축표시 170,180,190,200,210
# bar 바로선 막대그래프, barh 옆으로 막대그래프
plt.barh(x,y,label='키')
plt.legend(loc='upper center')  # 위치지점 변경

for i,txt in enumerate(y):
    plt.text(y[i]+1.8,x[i],txt,ha='center',color='blue')

plt.xticks([165,170,180,190,200,210])    # 그래프 y축 눈금표시
plt.xlim(165,210) # 그래프 눈금 범위 지정
plt.title('키 그래프')
```
