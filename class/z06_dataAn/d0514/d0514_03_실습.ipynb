```python
# chipotle.tsv 에서 주문중 가장 많이 주문한 메뉴 10개를 꺽은선 그래프로 출력하시오.
import pandas as pd
df = pd.read_csv('../z20_data/chipotle.tsv',sep='\t')
# df['item_price'] = df['item_price'].str.replace("$","").astype(float)
df

df['item_name'].value_counts(ascending=False)[:10]
df['item_name'].value_counts(ascending=False)[:10].index
df['item_name'].value_counts(ascending=False)[:10].values
```

```python
df.columns
```

```python
# item_name 별 주문한 개수의 합 출력하시오.
# 주문별 그룹
# quantity의 합을 구해야 함.
# df.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head()
df.groupby('item_name')['quantity'].sum().sort_values(ascending=False)[:5]
x = df.groupby('item_name')['quantity'].sum().sort_values(ascending=False)[:5].index
y = df.groupby('item_name')['quantity'].sum().sort_values(ascending=False)[:5].values
```

```python
# item_name 별 주문한 개수의 합 출력하시오.
# item_name별 주문한 횟수가 몇번인지 출력하시오.
df['item_name'].value_counts()[:5]
df['item_name'].value_counts()[:5].index
y2 = df['item_name'].value_counts()[:5].values
```

```python
plt.plot(x,y,label='주문수')
plt.plot(x,y2,label='주문횟수')
plt.xticks(rotation=90)
plt.legend(loc='upper right')
```
